```python
import numpy
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import find_peaks, peak_prominences
```

```python
aAB = 40
fA = 0.1
for x in range(13):
    filename = 'rep'+str(aAB)+'frac'+str(fA)+'/sq00'+str(x).zfill(2)+'.txt'
    data = numpy.loadtxt(filename)
    SQ = pd.DataFrame(data=data,columns=['q','Sq'])
    peaks, _ = find_peaks(SQ.Sq,distance=5)
    prominences = peak_prominences(SQ.Sq, peaks)[0]
    contour_heights = SQ.Sq[peaks] - prominences
    plt.plot(SQ.q,SQ.Sq)
    plt.plot(SQ.q[peaks],SQ.Sq[peaks],"x")
    plt.vlines(x=SQ.q[peaks], ymin=contour_heights, ymax=SQ.Sq[peaks])
    plt.xlabel('S(q)')
    plt.ylabel('q')
    plt.show()
```
