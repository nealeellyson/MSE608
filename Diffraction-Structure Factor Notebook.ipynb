```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import pandas as pd
from scipy.signal import find_peaks, peak_prominences
```

### Explore the full structure factor results from diffraction program for a given repulsion parameter and volume fraction at a variety of angles

```python
aAB = 65
phiB = 0.5
for x in range(13):
    filename = 'diffract/rep'+str(aAB)+'frac'+str(phiB)+'/sq00'+str(x).zfill(2)+'.txt'
    data = np.loadtxt(filename)
    SQ = pd.DataFrame(data=data,columns=['q','Sq'])
    peaks, _ = find_peaks(SQ.Sq,distance=5)
    prominences = peak_prominences(SQ.Sq, peaks)[0]
    contour_heights = SQ.Sq[peaks] - prominences
    plt.title("$a_{AB}$"+f"={aAB} &"" $\phi_{B}$"+f"={phiB}")
    plt.plot(SQ.q,SQ.Sq)
    plt.plot(SQ.q[peaks],SQ.Sq[peaks],"x")
    plt.vlines(x=SQ.q[peaks], ymin=contour_heights, ymax=SQ.Sq[peaks])
    plt.ylabel('S(q)')
    plt.xlabel('q')
    plt.show()
```

### Let's look at these displayed in their places on the phase diagram
What patterns do we see?

```python
imgs = [mpimg.imread(name) for name in imgs_names]
morph = ['S/C','C'  ,'3D/L'   ,'L' ,'L', # Here's some morphology suggestions/guesses
         'S'  ,'C'  ,'L'      ,'L' ,'L',
         'S/C','C'  ,'C/3D/PL','L' ,'L',
         'S'  ,'C'  ,'C/3D'   ,'L' ,'L',
         'S'  ,'C'  ,'C/3D'   ,'L' ,'L',
         'S'  ,'S/C','C'      ,'PL','L',
         'D'  ,'S'  ,'C'      ,'PL','L',
         'D'  ,'D'  ,'C'      ,'PL','L',
         'D'  ,'D'  ,'D'      ,'D' ,'D']
n_rows = 9 # Repulsion Parameter, aAB
n_cols = 5 # Volume Fraction, phiB
fig, axes = plt.subplots(n_rows, n_cols, figsize=[25,50])

for j in range(n_cols):
    
    for i in range(n_rows):
        axes[i][j].imshow(imgs[i*n_cols + j])
        axes[i][j].axis('off')
        axes[i][j].set_title(morph[i*n_cols + j]+": $a_{AB}$"+f"={65-i*5} &"" $\phi_{B}$"+f"=0.{j+1}")
plt.show()
```
